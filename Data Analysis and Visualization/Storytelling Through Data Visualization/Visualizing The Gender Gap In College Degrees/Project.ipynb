# Storytelling Through Data Visualization
## Visualizing The Gender Gap In College Degrees

### Import Statements

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Stylistics

```python
%matplotlib inline
plt.style.use("dark_background")
pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)
pd.set_option("display.width", 1000)
```

### Importing the Data Set

```python
women_degrees = pd.read_csv('../percent-bachelors-degrees-women-usa.csv')
```

### Data Exploration

```python
women_degrees.head()
```

```python
women_degrees.tail()
```

The first two cells are displayed to get familiarized about the data imported. Upon looking at the DataFrame, we can observe the following:
1. The `Year` column gives a natural ordering to the data set
2. The `Year` column span a values of 1970 - 2011
3. The rest of the column are the majors
4. All of the values of the column are float dtype (except Year)

```python
women_degrees.info()
```

Looking at this information, we are right to conclude that all of the values of the columns are in the float dtype (except Year). We can still conclude that:
1. We have 42 rows (0 to 41)
2. We have 18 columns
3. There are no nulls in our data set

```python
women_degrees.describe()
```

Looking at this information, we can conclude that the values of our columns are in percentages. Let's visualize some of the columns to further understand the data.

### Analysis

```python
dark_blue = (0/255,107/255,164/255)
orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

We can see in the plot that most of the degrees dominated by men, suggesting that there really is a gender gap.

Let's know visualize all of the 18 columns in the data set.

```python
fig = plt.figure(figsize=(16, 25))

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

for sp in range(0, 18, 3):
    idx = int(sp/3)
    ax = fig.add_subplot(6, 3, sp+1)

    ax.plot(women_degrees["Year"], women_degrees[stem_cats[idx]], c=dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[stem_cats[idx]], c=orange, linewidth=3, label="Men")
    ax.set_xlim(1968, 2000)
    ax.set_ylim(0, 100)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(stem_cats[idx])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

    if idx == 0:
        ax.text(1996, 80, 'Men')
        ax.text(1996, 10, 'Women')
    elif idx == 5:
        ax.text(1997, 88, 'Men')
        ax.text(1995, 5, 'Women')
        
for sp in range(1, 16, 3):
    idx = int((sp-1)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[idx]], c=dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[lib_arts_cats[idx]], c=orange, linewidth=3, label="Men")
    ax.set_xlim(1968, 2000)
    ax.set_ylim(0, 100)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(lib_arts_cats[idx])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if idx == 0:
        ax.text(1996, 77, 'Men')
        ax.text(1995, 15, 'Women')
        
for sp in range(2, 20, 3):
    idx = int((sp-2)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    
    ax.plot(women_degrees["Year"], women_degrees[other_cats[idx]], c=dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[other_cats[idx]], c=orange, linewidth=3, label="Men")
    ax.set_xlim(1968, 2000)
    ax.set_ylim(0, 100)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(other_cats[idx])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if idx == 0:
        ax.text(1996, 89, 'Men')
        ax.text(1995, 7, 'Women')
    elif idx == 5:
        ax.text(1997, 71, 'Men')
        ax.text(1995, 25, 'Women')

plt.show()
```
